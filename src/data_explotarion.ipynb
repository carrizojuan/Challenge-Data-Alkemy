```python
import os
import configparser
import requests
from datetime import datetime

#BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))

# Configuración credenciales
credentials_file = os.path.join("credentials", "access.conf")
parser = configparser.ConfigParser()
parser.read(credentials_file)

print(parser.get('postgre', 'user'))
```

```python
urls = {
    "museos": "https://datos.cultura.gob.ar/dataset/37305de4-3cce-4d4b-9d9a-fec3ca61d09f/resource/4207def0-2ff7-41d5-9095-d42ae8207a5d/download/museos_datosabiertos.csv",
    "cines": "https://datos.cultura.gob.ar/dataset/0560ef96-55ca-4026-b70a-d638e1541c05/resource/b8bf0459-16a5-430b-b8e1-4fb786572469/download/salas_cine.csv",
    "bibliotecas": "https://datos.cultura.gob.ar/dataset/37305de4-3cce-4d4b-9d9a-fec3ca61d09f/resource/01c6c048-dbeb-44e0-8efa-6944f73715d7/download/biblioteca_popular.csv"
}

# Obtener la ruta del directorio padre
directorio_padre = os.path.dirname(os.getcwd())  # Ruta del directorio actual

# Crear la ruta completa al directorio "data" en el directorio padre
ruta_data = os.path.join(directorio_padre, "data")

# Crear el directorio "data" en el directorio padre si no existe
os.makedirs(ruta_data, exist_ok=True)

today = datetime.now()


columns_norm_names=["cod_localidad", "id_provincia", "id_departamento","categoría",
              "provincia","localidad","nombre","domicilio","código postal","número de teléfono","mail","web"]
```

## MUSEOS

```python
import extract

path_museos = extract.create_filepath(today, "museos")
extract.extract(urls["museos"], path_museos)

```

```python
import pandas as pd

df_museos = pd.read_csv(path_museos)
column_rename = {
            'Cod_Loc':'cod_localidad',
            'IdProvincia':'id_provincia',
            'IdDepartamento':'id_departamento',
            'categoria':'categoría',
            'direccion':'domicilio',
            'CP':'código postal',
            'telefono':'número de teléfono',
            'Mail':'mail',
            'Web':'web'
}
df_museos.head()
```

```python
df_museos.rename(columns=column_rename, inplace=True)
df_museos=df_museos[columns_norm_names]
df_museos.info()
```

```python
df_museos.head()
```

## BIBLIOTECAS

```python
path_bibliotecas = extract.create_filepath(today, "bibliotecas")
extract.extract(urls["bibliotecas"], path_bibliotecas)
df_bibliotecas = pd.read_csv(path_bibliotecas)
df_bibliotecas.head()
```

```python
column_rename = {
            'categoria':'categoría',
            'cp':'código postal',
            'telefono':'número de teléfono'}
df_bibliotecas.rename(columns=column_rename, inplace=True)
df_bibliotecas=df_bibliotecas[columns_norm_names]
df_bibliotecas.head()
```

```python
df_bibliotecas.info()
```

```python
df_bibliotecas['número de teléfono'] = df_bibliotecas['número de teléfono'].astype(object)
df_bibliotecas['web'] = df_bibliotecas['web'].astype(object)
```

## CINES

```python
path_cines = extract.create_filepath(today, "cines")
#extract.extract(urls["cines"], path_cines)
df_cines = pd.read_csv(path_cines)
df_cines.head()
```

```python
column_rename = {
            'Cod_Loc':'cod_localidad',
            'IdProvincia':'id_provincia',
            'IdDepartamento':'id_departamento',
            'Categoría':'categoría',
            'Dirección':'domicilio',
            'CP':'código postal',
            'Teléfono':'número de teléfono',
            'Mail':'mail',
            'Web':'web',
            'Provincia':'provincia',
            'Localidad':'localidad',
            'Nombre':'nombre'}
df_cines.rename(columns=column_rename, inplace=True)
df_cines=df_cines[columns_norm_names]
df_cines.head()
```

```python
df_cines.info()
```

```python
df_cines['código postal'] = df_cines['código postal'].astype(object)
```

## MERGE DATA

```python
df_merge = pd.concat([df_bibliotecas, df_cines, df_museos], axis=0)
df_merge.tail()
```

### COUNT BY CATEGORY

```python
df_merge.groupby('categoría', as_index=False).size()
```

### COUNT BY SOURCE

```python
lst = list()
dfs = {"bibliotecas": df_bibliotecas, "cines": df_cines, "museos": df_museos}
for name, df in dfs.items():
    lst.append({'source': name, 'count':df.size})
pd.DataFrame(lst)
```

```python
df_merge.groupby(["provincia", "categoría"], as_index=False).size()
```
